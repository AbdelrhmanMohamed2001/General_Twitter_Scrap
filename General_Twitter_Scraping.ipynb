```python
import pandas as pd
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter

limit = 100
num_queries = int(input("Enter the number of queries to search for: "))

queries = []
for i in range(num_queries):
    query = input(f"Enter query #{i+1}: ")
    queries.append(query)

for query in queries:
    tweets = []

    for tweet in tqdm(sntwitter.TwitterSearchScraper(query).get_items(), desc=f"Collecting tweets for '{query}'"):
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.content, query])

    df = pd.DataFrame(tweets, columns=['Content', 'Query'])
    filename = f"{query.replace(' ', '_')}.csv"
    df.to_csv(filename, index=False)

    print(f"{len(tweets)} tweets collected for '{query}'. Saved to '{filename}'")
```
